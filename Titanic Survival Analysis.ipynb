# Data Cleaning

Data cleaning is a crucial step in the data analysis process for several reasons:

* Handling Missing Values: Real-world datasets often have missing data points. These can cause errors or biases in your analysis and modeling. We address missing values by techniques like filling them with a central tendency (mean, median, mode) or removing rows/columns, depending on the situation.

* Dealing with Inconsistent Data: Data might be recorded in different formats or have typos and inconsistencies. Cleaning helps standardize the data and ensure uniformity.

* Removing Duplicates: Duplicate entries can skew results. Identifying and removing duplicates is important for accurate analysis.

* Handling Outliers: Extreme values (outliers) can significantly impact statistical analysis and machine learning models. We identify and decide how to handle outliers (e.g., removing them, transforming the data) based on the context.

* Improving Data Quality for Analysis: Clean data is reliable data. By cleaning, we ensure that our analysis and any subsequent modeling are based on accurate and consistent information, leading to more trustworthy insights and predictions.

In summary, data cleaning ensures the data is accurate, consistent, and ready for analysis, leading to more reliable and meaningful results.

---


* .isna(): This method is used to detect missing values (represented as NaN, None, or NaT) in a DataFrame or Series. It returns a boolean DataFrame or Series of the same shape, where True indicates a missing value and False indicates a non-missing value. Think of it as asking, "Is this value missing?".

* .sum(): When applied to a boolean Series or DataFrame (like the one returned by .isna()), the .sum() method counts the number of True values along a specified axis. In pandas, True is treated as 1 and False as 0 in numerical contexts.

```python
# finding null val
titanic_df.isna().sum()
```

This tells us that:

* 'Age' has 177 missing values.

* 'Cabin' has 687 missing values.

* 'Embarked' has 2 missing values.

* All other columns have no missing values.

---

1. Filling with Median : For the 'Age' column, which is numerical, we filled the missing values with the median age of the existing data. This is a common strategy for numerical data as the median is less sensitive to outliers than the mean.

```python
# median of Age
Age_median = titanic_df['Age'].median()
print(Age_median)
```

```python
# replacing null val in Age with median
titanic_df['Age'] = titanic_df['Age'].fillna(Age_median)

# after Age cleaning
titanic_df[['Age','Cabin','Embarked']].isna().sum()
```

2. Filling with Mode : For the 'Embarked' column, which is categorical, we filled the missing values with the mode (the most frequent value). This is a suitable approach for categorical data where you want to replace missing values with the most common category.

```python
# mode of Embarked
Embarked_mode = titanic_df['Embarked'].mode()[0]
print(Embarked_mode)
```

```python
# replaing null val in Embarked with mode
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(Embarked_mode)

# after Embarked cleaning
titanic_df[['Age','Cabin','Embarked']].isna().sum()
```

3. Creating a New Feature and Dropping the Original : For the 'Cabin' column, which had a large number of missing values, we decided that simply filling them might not be informative. Instead, we created a new binary column called 'HasCabin' to indicate whether a passenger had a cabin or not (1 if they did, 0 if they didn't). We then dropped the original 'Cabin' column since the new feature captures the relevant information for our analysis.



```python
# creating new 'HasCabin' coloumn
titanic_df['HasCabin'] = titanic_df['Cabin'].notna().astype(int) # 1 if has cabin, 0 if not

titanic_df['Cabin'].value_counts()

display(titanic_df[['Cabin','HasCabin']])

# dropping 'Cabin' coloumn as its not needed
titanic_df.drop('Cabin', axis=1, inplace=True) # drop the 'Cabin' coloumn

# after Cabin coloumn replacement
titanic_df[['Age','HasCabin','Embarked']].isna().sum()

titanic_df['HasCabin'].value_counts()
```

```python
# 3. Extract 'Title' from the 'Name' column
titanic_df['Title'] = titanic_df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Let's see the different titles
print("Extracted Titles:")
titanic_df['Title'].value_counts()
```

```python
# Simplify the titles by grouping rare ones into a 'Rare' category
titanic_df['Title'] = titanic_df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

titanic_df['Title'] = titanic_df['Title'].replace('Mlle', 'Miss')
titanic_df['Title'] = titanic_df['Title'].replace('Ms', 'Miss')
titanic_df['Title'] = titanic_df['Title'].replace('Mme', 'Mrs')

# Let's see the survival rate by the new, cleaned titles
plt.figure(figsize=(12, 6))
sns.barplot(x='Title', y='Survived', data=titanic_df)
plt.title('Survival Rate by Title')
plt.ylabel('Survival Probability')
plt.show()
```

**Survival Rate by Title :**

This bar plot shows the survival rate based on the passenger's title.

*   **Key Insight :** Titles like 'Mrs.' and 'Miss' (which often indicate female passengers) have higher survival rates, while titles like 'Mr.' and 'Rare' (which includes many male titles) have lower survival rates.

# ydata Profiling

`ydata-profiling` (formerly `pandas-profiling`) is an open-source Python library that automates and simplifies the process of exploratory data analysis (EDA). It generates comprehensive and interactive HTML reports for your pandas DataFrames.

Instead of writing manual code to calculate descriptive statistics, visualize distributions, and identify missing values or correlations, `ydata-profiling` does all of this with a single line of code.

The report typically includes:

*   **Overview :** Essential statistics like the number of variables, observations, missing values, and duplicates.
*   **Variable types :** Identification of different variable types (numerical, categorical, date, etc.).
*   **Descriptive statistics :** Mean, median, standard deviation, min, max, quartiles, etc., for numerical variables. Value counts and frequencies for categorical variables.
*   **Visualizations :** Histograms, bar plots, and distribution plots to show the shape and spread of data.
*   **Interactions :** Scatter plots or other visualizations to show relationships between pairs of variables.
*   **Correlations :** Heatmaps or matrices to visualize correlations between numerical variables.
*   **Missing values :** Counts and visualizations of missing data patterns.
*   **Duplicate rows :** Identification of duplicate rows in the dataset.

In essence, `ydata-profiling` provides a quick and thorough way to get a deep understanding of your data, which is crucial before proceeding with further analysis or machine learning.
